```python
for i in House_sales.select_dtypes(include=["object"]).columns:
  print(i,House_sales[i].unique())
```

```python
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"Anna Nagar" if (i == "Ana Nagar" or i == "Ann Nagar") else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"Karapakkam" if (i == "Karapakam" ) else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"Adyar" if (i == "Adyr") else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"Velachery" if (i == "Velchery") else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"Chrompet" if (i == "Chrompt" or i == "Chrmpet" or i =="Chormpet") else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"KK Nagar" if (i == "KKNagar") else i)
House_sales['AREA'] = House_sales["AREA"].apply(lambda i:"T Nagar" if (i == "TNagar") else i)
```

```python
House_sales['SALE_COND'] = House_sales["SALE_COND"].apply(lambda i:"AbNormal" if (i == "Ab Normal") else i)
House_sales['SALE_COND'] = House_sales["SALE_COND"].apply(lambda i:"Partial" if (i == "Partiall" or i == "PartiaLl") else i)
House_sales['SALE_COND'] = House_sales["SALE_COND"].apply(lambda i:"AdjLand" if (i == "Adj Land") else i)
```
